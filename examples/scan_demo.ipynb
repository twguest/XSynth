```python
%load_ext autoreload
%autoreload 2

import numpy as np
from functools import partial

from xsynth.main import Scan
from xsynth.device import ADAPTX

scan = Scan(
    devices=[ADAPTX],
    scan_vectors=[np.linspace(-1.0, 1.0, 21)],
    beam_region = 'SA1',
    oscillators=["line"],
    oscillator_variables=[
        {
            "V2": 0.0,   # initial line offset
        }
    ],
    scan_variables=["V2"],
    wait_time=0.1,
    write=True,
    all_messages=True,
    display=False,
)

from xsynth.plot import plot_scan_summary

fig, axes = plot_scan_summary(scan)
```
