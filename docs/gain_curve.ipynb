```python
from xsynth.main import SetKicker
import numpy as np
import os
from xsynth.device import (KL2005, KMX1938, KMX1965, KNY1938, KNY1965)

```

```python
import pydoocs
import time

kick_x = np.linspace(-0.1,0.25,7)
kick_y = np.linspace(-0.1,0.25,7)

kicks = np.meshgrid(kick_x, kick_y)
scan_points = [pair for pair in zip(kicks[0].flatten(),kicks[1].flatten())]

pydoocs.write('XFEL.FEL/WAVELENGTHCONTROL.SA2/XFEL.SA2.COLOR1/CMD', 1)
time.sleep(15)

for scan_point in scan_points[:2]:
	K1, K2 = scan_point
	print(K1,K2)
	SetKicker(KMX1965,value = K1, start_time = 750, end_time=900)
	SetKicker(KNY1965,value = K1, start_time = 750, end_time=900)
	#script_call = "python ./gain_v2.py SASE2 10 32 /System/Volumes/Data/home/xfeloper/user/guestt/FD_09-11-24/gain_curve/{:.2f}V_{:.2f}V_".format(K1,K2)
	#os.system(script_call)	

	#pydoocs.write('XFEL.FEL/WAVELENGTHCONTROL.SA2/XFEL.SA2.COLOR1/CMD', 1)
	time.sleep(15)
```
